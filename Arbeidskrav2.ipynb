```python
import numpy as np

def fun_areal(a,b):
    
    # finn areal av figur
    arealTrekant = (a*b)/2
    r = a/2
    arealHalvsirkel = (np.pi*r**2)/2
    totAreal = arealTrekant + arealHalvsirkel
    
    # finn omkrets av figur
    finnHypotenus = np.sqrt(a**2+b**2)
    omkretsTrekant =finnHypotenus + a + b
    omkretsHalvsirkel = (2*np.pi*r)/2
    totOmkrets =omkretsTrekant + omkretsHalvsirkel
    
    return(totAreal,totOmkrets)


lengde = int(input('Oppgi lengden på trekant'))
høyde = int(input('Oppgi høyde på trekant')) 

(arealFigur,omkretsFigur) = fun_areal(høyde,lengde)


print('Arealet for figuren er:',arealFigur)
print('Omkretsen for figuren er:',omkretsFigur) 

```

```python
import numpy as np
import matplotlib.pyplot as plt

varX = int(input('Hvilket tall skal utgjøre x :'))

def fun_plot(varX):

    x = np.linspace(-varX,varX,num=200)

    y = -(x*x)-5 

    return (x,y)


(x_value,y_value) =fun_plot(varX)


plt.plot(x_value,y_value,color='darkGreen',linestyle=':')
plt.xlabel("x verdier")
plt.ylabel("y verdier")
plt.grid()
```
